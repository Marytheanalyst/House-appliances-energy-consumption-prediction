```python
X = mydata['T2'].values.reshape(-1, 1)
y = mydata['T6'].values

model = LinearRegression()
model.fit(X, y)

r2 = model.score(X, y)

print("R-squared value:", round(r2, 2))
```

```python
scaler = MinMaxScaler()
new_data = pd.DataFrame(scaler.fit_transform(mydata), columns = mydata.columns)
new_data.head()
```

```python
X = new_data.drop("Appliances", axis = 1)
Y = new_data["Appliances"]
```

```python
print(X.head())
print(Y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predicted_data = model.predict(x_test)
predicted_data
```

```python
# RSS
rss = np.sum(np.square(y_test - predicted_data))
round(rss, 2) 
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_data)
round(mae, 3)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_data))
round(rmse, 3) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_data)
round(r2_score, 2) 
```
